In this exercise, I analyze data from the prostate data set from the R package faraway. The prostate data frame has 97 rows and 9 columns and comes from a study on 97 men with prostate cancer who were due to receive a radical prostatectomy.
I use 2 variables in my analysis: 'lpsa' which represents the log(prostate specific antigen) and 'lcavol' which represents log(cancer volume).

The source of this data set is: Andrews DF and Herzberg AM (1985): Data. New York: Springer-Verlag

In the first part of this exercise, I perform a simple linear regression with lpsa as the reponse variable and lcavol as the predictor variable and I find the least square estimates for the coefficients. I also calculate a 95% confidence interval for both coefficients.

```r
library(faraway)
data(prostate)
Y <- prostate$lpsa # Response
X <- prostate$lcavol # Predictor
n <- length(X)

lm <- lm(Y~X)
(lms <- summary(lm))

coefficients <- c('beta0', 'beta1')

beta0.hat <- coef(lms)[1]
beta1.hat <- coef(lms)[2]
beta.hat <- c(beta0.hat, beta1.hat)

txx <- sum((X - mean(X))^2)
MSE <- sum(lms$residuals^2)/(n-2)

# Calculate standard deviations
s.beta0.hat <- sqrt(MSE * (1/n + mean(X)^2/(txx)))
s.beta1.hat <- sqrt(MSE/txx)
s.beta.hat <- c(s.beta0.hat, s.beta1.hat)

Yhat <- beta0.hat + beta1.hat * X

lower.bound <- beta.hat - qt(1 - 0.05/2, df = n-2)*s.beta.hat
upper.bound <- beta.hat + qt(1 - 0.05/2, df = n-2)*s.beta.hat

(beta.intervals <- cbind(coefficients, lower.bound, beta.hat, upper.bound))
```

In the second part of this exercise, I nominate potential outliers by singling out the externally studentized residuals that are outside the 95% confidence region of the Student's t distribution with 94 df. Potential outliers are represented by red squares in the plot below.

```r
SSE <- sum(lms$residuals^2)
hi <- 1/n+((X-mean(X))^2)/txx

#Calculate externally studentized residuals (residuals.star)
residuals.star <- lms$residuals*((n-3)/(SSE*(1-hi)-lms$residuals^2))^.5

plot(X, Y, col=ifelse (abs(residuals.star) >= qt(1-.05/2,n-3),"red","black"), pch=ifelse (abs(residuals.star) >= qt(1-.05/2,n-3),15,1), main = "lpsa v lcavol")
abline(lm)
```
