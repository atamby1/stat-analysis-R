In this exercise, I analyze data from the pipeline data set from the faraway R package.

The following is a description of the data set taken from https://cran.r-project.org/web/packages/faraway/faraway.pdf: 
"Researchers at National Institutes of Standards and Technology (NIST) collected data on ultrasonic measurements of the depths of defects in the Alaska pipeline in the field. The depth of the defects were then remeasured in the laboratory. These measurements were performed in six different batches. The laboratory measurements are more accurate than the in-field measurements, but more time consuming and expensive."

The pipeline data frame contains 107 observations on the following 3 variables. 'Field' is the measurement of depth of defect on site, 'Lab' is the measurement of depth of defect in the lab and 'Batch' is the batch of measurements.

This data comes from the Office of the Director of the Institute of Materials Research (now the Materials Science and Engineering Laboratory) of NIST.

```r
library(faraway)
data(pipeline)

Y <- pipeline$Lab
X <- pipeline$Field
n <- length(X)

lm <- lm(Y~X)
(lms <- summary(lm))

coefficient <- c('beta0', 'beta1')

beta0.hat <- (coef(lms)[1])
beta1.hat <- (coef(lms)[2])
beta.hat <- c(beta0.hat, beta1.hat)

txx <- sum((X - mean(X))^2) #Total Sum of Squares
MSE <- sum(lms$residuals^2)/(n-2) #Mean Squared Error

# Calculate standard deviations
s.beta0.hat <- sqrt(MSE * (1/n + mean(X)^2/(txx)))
s.beta1.hat <- sqrt(MSE/txx)
s.beta.hat <- c(s.beta0.hat, s.beta1.hat)

Yhat <- beta0.hat + beta1.hat * X

lower.bound <- beta.hat - qt(1 - 0.05/2, df = n-2)*s.beta.hat
upper.bound <- beta.hat + qt(1 - 0.05/2, df = n-2)*s.beta.hat

(beta.intervals <- cbind(coefficient, lower.bound, beta.hat, upper.bound))
plot(X,Y)
abline(lm)
```

From the residual plot below, it appears that the data is heteroscedastic, so I should try to find an appropriate variance-stabilizing transformation on one of the variables so that the relationship between the variables is approximately linear with constant variance.

```r
plot(X, resid(lm), main = 'residuals')
```

I try a log-log transformation and notice that the variables have an approximately linear relationship with constant variance, so I can now fit a linear regression for this model.

```r
Y <- log(pipeline$Lab)
X <- log(pipeline$Field)
n <- length(X)

lm2 <- lm(Y~X)
(lms2 <- summary(lm2))
plot(X,Y, main='log-log plot')
plot(X, resid(lm2), main = 'log-log residuals')
```
